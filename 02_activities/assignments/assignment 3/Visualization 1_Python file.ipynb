```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_document = 'https://ckan0.cf.opendata.inter.prod-toronto.ca/dataset/e28bc818-43d5-43f7-b5d9-bdfb4eda5feb/resource/01dff98a-b56b-4237-bb5d-f56319f0ced6/download/Marriage%20Licence%20Statistics%20Data.csv'
df = pd.read_csv(data_document)
```

```python
monthly_total = df.groupby('TIME_PERIOD')['MARRIAGE_LICENSES'].sum().reset_index()
x = monthly_total['TIME_PERIOD']
y = monthly_total['MARRIAGE_LICENSES']

fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(x,y, marker='o')

font1 = {'family': 'Times New Roman', 'color':'Black',
         'size':24}
font2 = {'family':'Arial', 'color':'indigo',
         'size':18}

plt.title('Total Marriage Licenses Issued Over Time in Toronto', fontdict = font1)
plt.xlabel('Date', fontdict = font2)
plt.ylabel('Number of Licenses', fontdict = font2)
plt.xticks(ticks=range(0, len(monthly_total), 6), rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```
